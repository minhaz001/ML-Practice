```python
x=np.arange(-4*np.pi,4*np.pi, step= 0.1)
y= np.sin(x)
```

```python
plt.title('Sine Curve')
plt.plot(x,y)
```

```python
#Subplot()
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r*--', linewidth= 0.5, markersize= 1) 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'go--', linewidth= 0.5, markersize= 1) 
plt.title('Cosine')  

plt.show()
```
