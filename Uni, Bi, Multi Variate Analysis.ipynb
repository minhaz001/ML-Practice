## Univariate Analysis

```python
## Separating each species
df_setosa=df.loc[df['species']=='setosa']
```

```python
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.show()
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o', color='m')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o', color='y')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o',color='k')
plt.xlabel('sepal length')
plt.show()
```

## Bivariate Analysis

```python
sns.FacetGrid(df,hue='species', size= 6).map(plt.scatter,'petal_length', 'sepal_width').add_legend()
plt.show()
```

```python
sns.jointplot(x='petal_length', y= 'sepal_width', data= df, kind='scatter', hue= 'species')
```

## Multivariate Analysis

```python
sns.pairplot(df, hue='species', size=3)
```
